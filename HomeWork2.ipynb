1. Вероятность того, что стрелок попадет в мишень, выстрелив один раз, равна 0.8. Стрелок выстрелил 100 раз. Найдите вероятность того, что стрелок попадет в цель ровно 85 раз.

```python
from math import factorial
# Создаем функцию для формул сочетания
def combinations (n, k):
    return int(factorial(n) // (factorial(k) * factorial(n - k)))
```

```python
p = 0.8
k = 85
n = 100
q = 1 - p

#Рассчитаем сочетания
с = combinations (n, k)

#Найдем вероятность того, что стрелок попадет в цель ровно 85 раз
Pn = с * p**k * q**(n-k)
print(Pn)
```

Ответ: 0.048 или 4.8%

3. Монету подбросили 144 раза. Какова вероятность, что орел выпадет ровно 70 раз?

```python
n5 = 144
k5 = 70
#Рассчитаем сочетания
с5 = combinations (n5, k5)

p5 = 0.5
q5 = 1 - p5

#Найдем вероятность того, что орел выпадет ровно 70 раз
Pn5 = с5 * p5**k5 * q5**(n5-k5)
print(Pn5)
```

Ответ:
Вероятность, что орел выпадет ровно 70 раз = 0.063 или 6,3%

4. В первом ящике находится 10 мячей, из которых 7 - белые. 
Во втором ящике - 11 мячей, из которых 9 белых. 
Из каждого ящика вытаскивают случайным образом по два мяча. 
Какова вероятность того, что все мячи белые? 
Какова вероятность того, что ровно два мяча белые?
Какова вероятность того, что хотя бы один мяч белый?

```python
#Найдем вероятность того, что все мечи белые

# Рассчитаем общее количество сочетаний 4 из 21
c1 = combinations (21, 4)
# Рассчитаем количество благоприятных исходов 4 из 16
c2 = combinations (16, 4)

P_o = c2/c1
print(round(P_o, 5))


#Найдем вероятность того, что ровно два мяча белые
n_1 = 10
k_1 = 7

n_2 = 11
k_2 = 9


P_a = k_1/n_1
P_b = k_2/n_2

P_ab = P_a * P_b
print(round(P_ab, 5))


#Найдем вероятность того, что хотя бы 1 мяч - белый

# Рассчитаем общее количество сочетаний 1 из 21
c_1 = combinations (21, 1)
# Рассчитаем количество благоприятных исходов 1 из 16
c_2 = combinations (16, 1)

P_1 = c_2/c_1
print(round(P_1, 5))
```

Ответ: 
Вероятность того, что все мячи белые = 0.30 или 30%
Вероятность того, что ровно два мяча белые = 0.57 или 57%
Вероятность того, что хотя бы один мяч белый = 0.76 или 76%
